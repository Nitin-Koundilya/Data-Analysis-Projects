-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

temp=new_movies['genres'].str.split("|",n=2,expand=True) #Splitting genres column against pipe(|) and creating two columns to store first two splitted terms in temp variable.

temp[1].fillna(value=temp[0],axis=0, inplace=True) # Imputing the null values of genre_2 by genre_1's values

new_movies['genre_1']=temp[0] #adding 1st column of temp variable as genre_1 in new_movies dataframe
new_movies['genre_2']=temp[1] #adding 2nd column of temp variable as genre_2 in new_movies dataframe
new_movies # cross checking the added values
```

```python
 # Write your code for grouping the dataframe here
movies_by_segment = new_movies.groupby(by=['genre_1','genre_2'])
```

```python
PopGenre = movies_by_segment.mean() #Applying mean function on groped dataframe

# Write your code for getting the 5 most popular combo of genres here

PopGenre=PopGenre.sort_values(by=['gross'],ascending=False).head(5) #Sorting grouped dataframe with reference to column mean value of gross and storing top 5 

PopGenre # Cross checking
```

```python
# Write the code for calculating decade here
#Adding new column decades in new_movies
new_movies['decade']=new_movies['title_year'].map(lambda x: int((x//10)*10)).astype(str)+'s'

new_movies[['title_year','decade']].tail(10) #Displaying last 5 values of decades and title_year after adding decades column for cross checking
```

```python
# Write your code for creating the data frame df_by_decade here 

df_by_decade=new_movies.sort_values(by='decade').groupby(by="decade").num_voted_users.sum() #Sorting dataframe new_movies by decade, groupomg by decade and applying sum on num_voted_userss and storing it in df_by_decade

df_by_decade #Checking 
```

```python
# Write your code for plotting number of voted users vs decade

plt.figure(figsize=(14,5)) # creating a new figure with its dimensions as 14*5
plt.xlabel('DECADE') #Naming x axis 
plt.ylabel('SUM OF USERS VOTED') # Naming y axis
plt.title("SUM OF USERS VOTED IN EACH DECADE") #Setting title of figure
plt.bar(df_by_decade.index,df_by_decade) #Plotting bar graph to depict sum of users voted in each decade
plt.show() #Displaying figure
```
